```python
import os
import json

import pandas as pd
import mysql.connector as mysql
from sqlalchemy import create_engine
```

```python
def state_name_format(state_name):
    state_name = state_name.replace("-"," ")
    return state_name.title()
```

```python
#This is to direct the path to get the data as states

Agg_Transaction_path="/Users/upendra/Desktop/Capstone_Projets/PhonePe/data/data/aggregated/transaction/country/india/state/"
Agg_states_list=os.listdir(Agg_Transaction_path)
#print(Agg_states_list)

#Agg_user_path = "/Users/upendra/Desktop/Capstone_Projets/PhonePe/data/data/aggregated/user/country/india/state/"
#Agg_users_list = os.listdir(Agg_user_path)
#print(Agg_users_list)
#This is to extract the data's to create a dataframe

Agg_states_data = {'State':[], 'Year':[], 'Quarter':[], 'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

def Agg_detail_states_list(state_list):
    for i in state_list:
        country_name = Agg_Transaction_path+i+"/"
        try:
            select_years_list = os.listdir(country_name)
            for j in select_years_list:
                year_path = country_name+j+"/"
                Agg_year_list = os.listdir(year_path)
                for k in Agg_year_list:
                    selected_year = year_path+k
                    data = open(selected_year,'r')
                    json_data = json.load(data)

                    transactionData = json_data['data']['transactionData']
                    #print("transactionData.........",transactionData)
                    for details in transactionData:
                        name = details['name']
                        count = details['paymentInstruments'][0]['count']
                        amount = details['paymentInstruments'][0]['amount']
                        #print('name......',name)

                        Agg_states_data['Transaction_type'].append(name)
                        Agg_states_data['Transaction_count'].append(count)
                        Agg_states_data['Transaction_amount'].append(amount)
                        Agg_states_data['State'].append(state_name_format(i))
                        Agg_states_data['Year'].append(j)
                        Agg_states_data['Quarter'].append(int(k.strip('.json')))
                        
                        
            Agg_Trans_df =pd.DataFrame(Agg_states_data)
            Agg_Trans_df = Agg_Trans_df.drop_duplicates()
        except:
            pass
        finally:
            data.close()
    return Agg_Trans_df

#Agg_detail_states_list(Agg_states_list)
```

```python
Agg_user_path = "/Users/upendra/Desktop/Capstone_Projets/PhonePe/data/data/aggregated/user/country/india/state/"
Agg_users_list = os.listdir(Agg_user_path)
#print(Agg_users_list)
```

```python
Agg_users_data = {'registeredusers':[],'State':[], 'Year':[], 'Quarter':[], 'Brand':[], 'Count':[], 'Percentage':[]}
```

```python
def Agg_user_list(user_list):
    for i in user_list:
        user_state = Agg_user_path + i+"/"
        #print('user_state.....', user_state)
        try:
            user_year_list = os.listdir(user_state)
            for j in user_year_list:
                #print('j....',j)
                user_year = user_state+j+"/"
                selected_year_list = os.listdir(user_year)
                for k in selected_year_list:
                    #print('k......',k)
                    selected_year = user_year+k
                    data = open(selected_year,"r")
                    user_json = json.load(data)

                    userData = user_json['data']['usersByDevice']
                    agg_userData = user_json['data']['aggregated']
                    #print('userdata.......',userData)

                    if userData:
                        for details in userData:
                            brand = details['brand']
                            count = details['count']
                            percentage = details['percentage']

                            Agg_users_data['Brand'].append(brand)
                            Agg_users_data['Count'].append(count)
                            Agg_users_data['Percentage'].append(percentage)
                            Agg_users_data['State'].append(state_name_format(i))
                            Agg_users_data['Year'].append(j)
                            Agg_users_data['Quarter'].append(int(k.strip('.json')))
                            Agg_users_data['registeredusers'].append(agg_userData['registeredUserss'])
                            
                    
            Agg_user_df = pd.DataFrame(Agg_users_data)

        except:
            pass
        
        finally:
            data.close()
            

    return Agg_user_df
                    
#Agg_user_df = Agg_user_list(Agg_users_list)
#print('agg user....',Agg_user)
```

```python
def Map_transaction_list(trans_data):
    #print('trans_data========',trans_data)
    for i in trans_data:
        #print('i......',i)
        trans_name = Map_trans_path+i+"/"
        try:
            trans_year_list = os.listdir(trans_name)
            for j in trans_year_list:
               # print('j......',j)
                trans_year = trans_name+j+"/"
                selected_year_list = os.listdir(trans_year)
                for k in selected_year_list:
                    #print('k.........',k)
                    selected_year = trans_year+k
                    data = open(selected_year,'r')
                    map_trans_json = json.load(data)
                    
                    map_transaction_data = map_trans_json['data']['hoverDataList']
                    #print('map_transaction_data........',map_transaction_data)
                    for details in map_transaction_data:
                        name = details['name']
                        Type = details['metric'][0]['type']
                        count = details['metric'][0]['count']
                        amount = details['metric'][0]['amount']
                        
                        Map_trans_data['Name'].append(name)
                        Map_trans_data['Type'].append(Type)
                        Map_trans_data['Count'].append(count)
                        Map_trans_data['Amount'].append(amount)
                        Map_trans_data['State'].append(state_name_format(i))
                        Map_trans_data['Year'].append(j)
                        Map_trans_data['Quarter'].append(int(k.strip('.json')))
                        
                        
            map_trans_df = pd.DataFrame(Map_trans_data)
        except:
            pass
        
        finally:
            data.close()
                  
                        
    return map_trans_df

#map_trans_details = Map_transaction_list(Map_trans_list)
#print('map_trans_details.......',map_trans_details)
```

```python
Map_Users_Data = {'State':[],'Year':[],'Quarter':[],'Name':[],'RegisteredUsers':[],'AppOpens':[]}
```

```python
def Map_users_list(map_users):
    for i in map_users:
        map_user_state = Map_user_path+i+"/"
        try:
            map_user_years = os.listdir(map_user_state)
            for j in map_user_years:
                user_years = map_user_state + j+"/"
                years_list =  os.listdir(user_years)
                for k in years_list:
                    selected_year =  user_years+k
                    data = open(selected_year,'r')
                    map_user_json = json.load(data)
                    
                    map_user_data = map_user_json['data']['hoverData'].items()
                    for details in map_user_data:
                        district = details[0]
                        registered_users = details[1]['registeredUsers']
                        app_opens = details[1]['appOpens']
                        
                        
                        Map_Users_Data['Name'].append(district)
                        Map_Users_Data['RegisteredUsers'].append(registered_users)
                        Map_Users_Data['AppOpens'].append(app_opens)
                        Map_Users_Data['State'].append(state_name_format(i))
                        Map_Users_Data['Year'].append(j)
                        Map_Users_Data['Quarter'].append(int(k.strip('.json')))
                        
            map_user_df = pd.DataFrame(Map_Users_Data)
                        
        except:
            pass
        
        finally:
            data.close()
            
                        
    return map_user_df

#map_user_details = Map_users_list(Map_user_list)
#print(map_user_details)
```

```python
Top_transaction_path = "/Users/upendra/Desktop/Capstone_Projets/PhonePe/data/data/top/transaction/country/india/state/"
Top_transaction_list = os.listdir(Top_transaction_path)
#print(Top_transaction_list)
```

```python
def load_top_trans_data():
        top_trans_district=[]
        top_trans_pincode=[]
        for state in Top_transaction_list:
            try:
                year_path=Top_transaction_path+state+'/'
                year_list=os.listdir(year_path)
                for year in year_list:
                    quarter_path=year_path+year+'/'
                    quarter_list=os.listdir(quarter_path)
                    for file in quarter_list:
                        file_path=quarter_path+file
                        file_opened=open(file_path,'r')
                        data=json.load(file_opened)
                        #print('data.......',data)

                        for district in data['data']['districts']:
                            quarter=int(file.strip('.json'))
                            top_trans_district.append({
                                'District':district.get('entityName','NA'),
                                'Count':district['metric'].get('count', 0),
                                'Amount':district['metric'].get('amount', 0),
                                'State':state_name_format(state),
                                'Year':year,
                                'Quarter':quarter
                            })

                        for pincode in data['data']['pincodes']:
                            quarter=int(file.strip('.json'))
                            top_trans_pincode.append({
                                'Pincode':pincode.get('entityName','NA'),
                                'Count':pincode['metric'].get('count', 0),
                                'Amount':pincode['metric'].get('amount', 0),
                                'State':state_name_format(state),
                                'Year':year,
                                'Quarter':quarter

                            })
            except:
                pass
            finally:
                file_opened.close()
                
            
        top_trans_district_df=pd.DataFrame(top_trans_district)
        top_trans_district_df['Amount']=top_trans_district_df['Amount'].apply(lambda x: round(x,2))
        top_trans_pincode_df=pd.DataFrame(top_trans_pincode)
        top_trans_pincode_df['Amount']=top_trans_pincode_df['Amount'].apply(lambda x: round(x,2))

        return top_trans_district_df,top_trans_pincode_df
    
    
```

```python
Top_users_path = "/Users/upendra/Desktop/Capstone_Projets/PhonePe/data/data/top/user/country/india/state/"
Top_users_list = os.listdir(Top_users_path)
#print(Top_users_list)
```

```python
def load_top_user_data():
    top_user_districts = []
    top_user_pincodes = []
    for state in Top_users_list:
        try:
            year_path = Top_users_path+state+"/"
            year_list = os.listdir(year_path)
            for year in year_list:
                quarter_path = year_path+year+"/"
                quarter_list = os.listdir(quarter_path)
                for quarter in quarter_list:
                    #print('quarter..........',quarter)
                    selected_quarter = quarter_path+quarter
                    
                    data = open(selected_quarter,'r')
                    top_user_json = json.load(data)
                    
                    for district in top_user_json['data']['districts']:
                        quarter_num=int(quarter.strip('.json'))
                        top_user_districts.append({
                                'District':district.get('name','NA'),
                                'RegisteredUsers':district.get('registeredUsers', 0),
                                'State':state_name_format(state),
                                'Year':year,
                                'Quarter':quarter_num      
                        })
                    
                    
                   # print('top_user_json pincodes.......',len(top_user_json['data']['pincodes']))
                    for pincode in top_user_json['data']['pincodes']:
                        #print(pincode)
                        quarter_num=int(quarter.strip('.json'))
                        top_user_pincodes.append({
                                'Pincode':pincode.get('name','NA'),
                                'RegisteredUsers':pincode.get('registeredUsers', 0),
                                'State':state_name_format(state),
                                'Year':year,
                                'Quarter':quarter_num      
                        })
        except Exception as err:
            print(err)
            pass
        finally:
            data.close()
        
    top_user_districts_df = pd.DataFrame(top_user_districts)
    top_user_pincodes_df = pd.DataFrame(top_user_pincodes)
    
    return top_user_districts_df,top_user_pincodes_df
```
